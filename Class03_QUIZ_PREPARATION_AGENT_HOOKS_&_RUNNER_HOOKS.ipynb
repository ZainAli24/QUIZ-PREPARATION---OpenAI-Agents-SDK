# **1. Agent Hook and RunHook Lifecycle:**

---

### **Hook Ka Matlab Kya Hai?**
Programming mein, **hook** ek aisa mechanism hota hai jo aap ko kisi system ke execution ke specific points par intervene (dakhal dena) ya observe (dekhna) karne ki ijazat deta hai. Yeh aksar event-driven programming mein use hota hai, jahan aap kisi event (jaise ke start hona, end hona) ke saath apna custom code attach kar sakte hain. OpenAI Agents SDK ke context mein, hooks ka matlab hai agent ke lifecycle ke douran hone wale events ko monitor karna ya unmein custom behavior add karna. Yeh lifecycle agent ke chalne ke mukhtalif stages ko kehta hai, jaise agent ka shuru hona, input process karna, tools use karna, ya output dena.

Hooks ka basic idea yeh hai ke aap agent ke behavior ko bina uske core logic ko change kiye customize kar sakte hain. For example, aap chahte hain ke jab agent start ho to ek log entry ban jaye, ya jab tool chalaye to uska result check ho—yeh sab hooks ke zariye possible hai.

---

### **OpenAI Agents SDK Mein Hooks**
OpenAI Agents SDK mein do main types ke hooks hain:
1. **`AgentHook`**: Yeh ek specific agent ke lifecycle events ke liye hota hai. Yani, aap kisi ek agent ke start, end, ya tool usage ko observe kar sakte hain.
2. **`RunHook`**: Yeh poore agent run ke lifecycle events ke liye hota hai, jo multiple agents ko involve kar sakta hai. Yani, yeh overall workflow ko dekhta hai.

Dono hooks ko aap subclass kar ke apne custom methods define kar sakte hain, aur phir unhe agent ya runner ke saath attach kar sakte hain.

---

### **AgentHook: Methods, Parameters, aur Kaam**
`AgentHook` ek class hai jo aap ko ek specific agent ke events ke liye callbacks provide karti hai. Aap is class ko subclass kar ke apne methods override kar sakte hain. Yeh raha is ke methods ka detail:

#### 1. **`on_start(context: RunContextWrapper[TContext], agent: Agent[TContext]) -> None`**
- **Purpose**: Yeh tab call hota hai jab agent process shuru hone wala hota hai.
- **Parameters**:
  - `context`: Ek `RunContextWrapper` object jo current run ke state ko represent karta hai (jaise conversation history ya shared data).
  - `agent`: Woh agent object jo start ho raha hai.
- **Kaise Kaam Karta Hai**: Is method mein aap agent ke shuru hone se pehle kuch kar sakte hain, jaise state initialize karna ya log karna.
- **Example Use**: "Agent XYZ shuru ho raha hai" print karna.

#### 2. **`on_end(context: RunContextWrapper[TContext], agent: Agent[TContext], output: Any) -> None`**
- **Purpose**: Yeh tab call hota hai jab agent apna final output de deta hai.
- **Parameters**:
  - `context`: Run ka context.
  - `agent`: Agent jo khatam hua.
  - `output`: Agent ka final output (koi bhi data type ho sakta hai).
- **Kaise Kaam Karta Hai**: Aap output ko log kar sakte hain, uspar post-processing kar sakte hain, ya koi action trigger kar sakte hain.
- **Example Use**: Output ko database mein save karna.

#### 3. **`on_handoff(context: RunContextWrapper[TContext], agent: Agent[TContext], source: Agent[TContext]) -> None`**
- **Purpose**: Yeh tab call hota hai jab ek agent se doosre agent ko control handoff hota hai.
- **Parameters**:
  - `context`: Run ka context.
  - `agent`: Naya agent jo control le raha hai.
  - `source`: Woh agent jis ne control diya.
- **Kaise Kaam Karta Hai**: Yeh agents ke beech flow ko track karne ke liye useful hai.
- **Example Use**: Handoff ka log banana, jaise "Agent A se Agent B ko handoff hua".

#### 4. **`on_tool_start(context: RunContextWrapper[TContext], agent: Agent[TContext], tool: Tool) -> None`**
- **Purpose**: Yeh tab call hota hai jab agent ek tool ko invoke karne wala hota hai.
- **Parameters**:
  - `context`: Run ka context.
  - `agent`: Agent jo tool use kar raha hai.
  - `tool`: Tool object jo chalaya ja raha hai.
- **Kaise Kaam Karta Hai**: Tool ke chalne se pehle aap kuch kar sakte hain, jaise tool ka naam log karna.
- **Example Use**: Tool ke input ko check karna.

#### 5. **`on_tool_end(context: RunContextWrapper[TContext], agent: Agent[TContext], tool: Tool, result: str) -> None`**
- **Purpose**: Yeh tab call hota hai jab tool invoke ho chuka hota hai aur result de deta hai.
- **Parameters**:
  - `context`: Run ka context.
  - `agent`: Agent jo tool use kar raha tha.
  - `tool`: Tool jo chala.
  - `result`: Tool ka result (string format mein).
- **Kaise Kaam Karta Hai**: Tool ke output ko monitor ya process karne ke liye.
- **Example Use**: Result ko validate karna ya log karna.

#### **Note**:
- `TContext` ek generic type hai, yani aap apne application ke specific context ko define kar sakte hain.
- Har method mein `context` se aap run ke current state tak pohonch sakte hain.

### **RunHook: Methods, Parameters, aur Kaam**

`RunHook` ek class hai jo agent run ke lifecycle events par callbacks provide karti hai. Yeh class aap ko poore run ko monitor aur customize karne ki flexibility deti hai, chahe usmein multiple agents involved hon. Aap isko subclass kar ke apne custom methods define kar sakte hain aur phir runner ke saath attach kar sakte hain.

### **RunHook Ke Methods**
Documentation ke mutabiq, `RunHook` mein yeh methods hain jo aap override kar sakte hain:

#### 1. **`on_agent_start` (async)**
- **Signature**:  
  ```python
  on_agent_start(context: RunContextWrapper[TContext], agent: Agent[TContext]) -> None
  ```
- **Purpose**: Yeh method tab call hota hai jab koi agent invoke hone wala hota hai. Har baar jab current agent change hota hai, yeh chalega.
- **Parameters**:
  - `context`: `RunContextWrapper[TContext]` - Current run ka state represent karta hai.
  - `agent`: `Agent[TContext]` - Wo agent jo start ho raha hai.
- **Kaise Kaam Karta Hai**: Agent ke shuru hone se pehle aap logging, state initialization, ya koi aur action kar sakte hain.
- **Example**:  
  ```python
  async def on_agent_start(self, context, agent):
      print(f"Agent {agent.name} invoke ho raha hai")
  ```

#### 2. **`on_agent_end` (async)**
- **Signature**:  
  ```python
  on_agent_end(context: RunContextWrapper[TContext], agent: Agent[TContext], output: Any) -> None
  ```
- **Purpose**: Jab agent apna final output de deta hai, tab yeh method call hota hai.
- **Parameters**:
  - `context`: Run ka context.

  - `output`: Agent ka final output (koi bhi data type ho sakta hai).
- **Kaise Kaam Karta Hai**: Output ko log karna, post-process karna, ya koi action trigger karna possible hai.
- **Example**:  
  ```python
  async def on_agent_end(self, context, agent, output):
      print(f"Agent {agent.name} ka output: {output}")
  ```

#### 3. **`on_handoff` (async)**
- **Signature**:  
  ```python
  on_handoff(context: RunContextWrapper[TContext], from_agent: Agent[TContext], to_agent: Agent[TContext]) -> None
  ```
- **Purpose**: Ek agent se doosre agent ko control handoff hone par yeh call hota hai.
- **Parameters**:
  - `context`: Run ka context.
  - `from_agent`: Agent jo control de raha hai.
  - `to_agent`: Agent jo control le raha hai.
- **Kaise Kaam Karta Hai**: Agents ke beech flow ko track karne ke liye useful.
- **Example**:  
  ```python
  async def on_handoff(self, context, from_agent, to_agent):
      print(f"Handoff from {from_agent.name} to {to_agent.name}")
  ```

#### 4. **`on_tool_start` (async)**
- **Signature**:  
  ```python
  on_tool_start(context: RunContextWrapper[TContext], agent: Agent[TContext], tool: Tool) -> None
  ```
- **Purpose**: Jab agent koi tool invoke karne wala hota hai, tab yeh method call hota hai.
- **Parameters**:
  - `context`: Run ka context.
  - `agent`: Agent jo tool use kar raha hai.
  - `tool`: `Tool` object jo chalaya ja raha hai.
- **Kaise Kaam Karta Hai**: Tool ke chalne se pehle aap uska naam log kar sakte hain ya input check kar sakte hain.
- **Example**:  
  ```python
  async def on_tool_start(self, context, agent, tool):
      print(f"Tool {tool.name} start ho raha hai by {agent.name}")
  ```

#### 5. **`on_tool_end` (async)**
- **Signature**:  
  ```python
  on_tool_end(context: RunContextWrapper[TContext], agent: Agent[TContext], tool: Tool, result: str) -> None
  ```
- **Purpose**: Tool invoke hone ke baad aur result dene par yeh call hota hai.
- **Parameters**:
  - `context`: Run ka context.
  - `agent`: Agent jo tool use kar raha tha.
  - `tool`: Tool jo chala.
  - `result`: Tool ka result (string format mein).
- **Kaise Kaam Karta Hai**: Tool ke output ko monitor ya process karne ke liye.
- **Example**:  
  ```python
  async def on_tool_end(self, context, agent, tool, result):
      print(f"Tool {tool.name} ka result: {result}")
  ```

### **RunHook vs. AgentHook**
- **`RunHook`**: Poore run ke lifecycle events ko cover karta hai, chahe kitne agents involved hon. Yeh runner level par attach hota hai.
- **`AgentHook`**: Sirf ek specific agent ke events ko cover karta hai aur agent level par attach hota hai.
- **Fark**: Methods same hain, lekin `RunHook` overall run ke liye hai, jabke `AgentHook` sirf us agent ke liye.

---

### **Kaise Use Karte Hain?**
Aap `RunHook` ko subclass karte hain, apne methods override karte hain, aur phir usko `Runner.run` method mein pass karte hain.

**Example Code**:
```python
from agents import Runner, RunHook, RunContextWrapper, Agent, Tool

class MyRunHook(RunHook[MyContext]):
    async def on_agent_start(self, context: RunContextWrapper[MyContext], agent: Agent[MyContext]):
        print(f"Agent {agent.name} start ho raha hai")
    
    async def on_agent_end(self, context: RunContextWrapper[MyContext], agent: Agent[MyContext], output: Any):
        print(f"Agent {agent.name} ka output: {output}")
    
    async def on_handoff(self, context: RunContextWrapper[MyContext], from_agent: Agent[MyContext], to_agent: Agent[MyContext]):
        print(f"Handoff from {from_agent.name} to {to_agent.name}")
    
    async def on_tool_start(self, context: RunContextWrapper[MyContext], agent: Agent[MyContext], tool: Tool):
        print(f"Tool {tool.name} start ho raha hai by {agent.name}")
    
    async def on_tool_end(self, context: RunContextWrapper[MyContext], agent: Agent[MyContext], tool: Tool, result: str):
        print(f"Tool {tool.name} ka result: {result}")

# Usage
runner = Runner()
result = await runner.run(agent, input, hooks=MyRunHook())
```

### **Summary**
- `RunHook` aap ko agent run ke har step (agent start/end, handoff, tool start/end) par control deta hai.
- Yeh methods asynchronous hain, to aap complex async operations bhi kar sakte hain.
- Parameters ke zariye aap ko context, agents, tools, aur results tak access milta hai.
- Custom logic add karne ka yeh ek powerful tareeka hai bina core agent logic ko change kiye.

Hooks ko use karne ka process yeh hai:
1. **Attach Karna**:
   - `AgentHook` ko agent ke `hooks` property ke zariye attach karte hain.
   - `RunHook` ko `Runner.run` method mein pass karte hain.
2. **Callbacks**:
   - Jab run ya agent ke lifecycle mein events hote hain, SDK automatically corresponding hook methods ko call karta hai.
3. **Custom Logic**:
   - Aap apne subclass mein methods override kar ke custom behavior define karte hain.

### **Summary**

- **`AgentHook`** specific agent ke events (start, end, handoff, tool usage) ke liye hai.
- **`RunHook`** poore run ke events (run start, end, agent switch) ke liye hai.
- Har method ke parameters (jaise `context`, `agent`, `output`) aap ko run ke state aur event details dete hain.

#### 1. **Asynchronous Code Ko Run Karna**
Python mein `asyncio` ek library hai jo **asynchronous programming** ke liye use hoti hai. Asynchronous programming ka matlab hai ke aapka code concurrently kaam kar sakta hai, khas tor par jab operations time-consuming hote hain (jaise network requests ya I/O operations).

Aapka `main` function `async def` se define kiya gaya hai, jo ek **coroutine** banata hai. Coroutines ko directly call nahi kiya ja sakta; unhe ek **event loop** ke zariye run karna padta hai. `asyncio.run` yeh kaam karta hai:

- **Event Loop Create Karta Hai**: `asyncio.run` ek naya event loop banata hai, jo asynchronous tasks ko schedule aur execute karta hai.
- **Coroutine Ko Run Karta Hai**: Yeh `main()` coroutine ko event loop mein run karta hai aur uske complete hone ka wait karta hai.
- **Event Loop Close Karta Hai**: Jab coroutine complete hota hai, `asyncio.run` event loop ko cleanup karta hai.

  ### Toh, `asyncio.run(main())` ka maqsad hai ke aapka `main` function, jo asynchronous hai, ko sahi tarike se execute kiya jaye.

### **1. `RunContextWrapper` Ka Basic Structure**
`RunContextWrapper` ek **generic class** hai jo Python ke `Generic[TContext]` ka use karti hai. Iska matlab hai ke yeh class kisi bhi type ke context ke saath kaam kar sakti hai, jab tak aap specify karte hain ke `TContext` kya hai. Yeh hai `RunContextWrapper` ki definition:
```python
@dataclass
class RunContextWrapper(Generic[TContext]):
    context: TContext
    usage: Usage = field(default_factory=Usage)
```
- **`context`**: Yeh attribute `TContext` type ka hai, jo aapke case mein `MyCustomContext` ban jata hai.
- **`usage`**: Yeh ek `Usage` type ka object hai jo agent ke run ke usage details (jaise tokens) ko track karta hai.

1. `Runner.run()` method aapke `my_context` ko leta hai.

Isliye, aapke `on_agent_start` method mein `context` ek `RunContextWrapper[MyCustomContext]` object hota hai, jo SDK ne banaya hai.

## **8. RunHook(Global Run level) and AgentHook(Specific Agent level):**

### **2. Kya `AgentHooks` `RunHooks` Ko Override Karta Hai?**
Ab aapka main sawal: agar aapne `Agent` mein `AgentHooks` define kiya aur `Runner.run()` mein `RunHooks` define kiya, toh kya `AgentHooks` us specific agent ke liye `RunHooks` ko override karega?

**Jawab**: **Haan, `AgentHooks` us specific agent ke liye `RunHooks` ko override karta hai**, lekin yeh override sirf us agent ke lifecycle events ke liye hota hai jiske liye `AgentHooks` define kiya gaya hai. Doosre agents ke liye, `RunHooks` kaam karta rahega. Yeh SDK ke design ke hisaab se hota hai. Documentation ke mutabiq:

- `RunHooks` global hooks hain jo **sab agents** ke events ke liye trigger hote hain, jab tak koi agent-specific `AgentHooks` define na ho.
- `AgentHooks` agent-specific hooks hain jo usi agent ke events ke liye trigger hote hain, aur agar yeh define hote hain, toh woh us agent ke liye `RunHooks` ke corresponding methods ko **override** kar dete hain.

---

### **3. Yeh Kaise Hota Hai? (SDK Ka Internal Logic)**
SDK ka internal logic yeh ensure karta hai ke `AgentHooks` ko precedence di jaye jab specific agent ke events trigger hote hain. Yeh kaise kaam karta hai, iska simplified explanation:

1. **Event Triggering**:
   - Jab `Runner.run()` chal raha hota hai, SDK lifecycle events (jaise `on_agent_start`, `on_tool_start`, `on_handoff`) ko track karta hai.
   - Har event ke liye, SDK check karta hai:
     - Kya current agent ke pas `hooks` parameter mein `AgentHooks` define hai?
     - Kya `Runner.run()` mein global `RunHooks` define hai?

2. **Precedence Logic**:
   - **Agar `AgentHooks` define hai**: SDK us agent ke `AgentHooks` ke methods ko call karta hai us specific agent ke events ke liye. `RunHooks` ke corresponding methods ko ignore kiya jata hai.
   - **Agar `AgentHooks` nahi hai**: SDK `RunHooks` ke methods ko call karta hai (agar yeh define hai).
   - **Agar dono nahi hain**: Koi hook call nahi hota, aur default behavior (empty `pass` methods) use hota hai.

3. **Handoffs aur Multiple Agents**:
   - Jab ek agent doosre agent ko handoff karta hai (jaise `Assistant` se `TripPlane_Agent`), SDK check karta hai ke handoff agent ke pas `AgentHooks` define hai ya nahi.
   - Agar handoff agent ke pas `AgentHooks` hai, toh uske events ke liye `AgentHooks` call hote hain. Warna, global `RunHooks` call hote hain.

**Pseudocode (Inferred)**:
```python
async def handle_lifecycle_event(event, agent, context_wrapper, tool=None, from_agent=None, to_agent=None):
    # Check if agent has AgentHooks
    if agent.hooks:
        if event == "agent_start":
            await agent.hooks.on_agent_start(context_wrapper, agent)
        elif event == "tool_start":
            await agent.hooks.on_tool_start(context_wrapper, agent, tool)
        # ... other events
    # Otherwise, fall back to RunHooks
    elif run_hooks:
        if event == "agent_start":
            await run_hooks.on_agent_start(context_wrapper, agent)
        elif event == "tool_start":
            await run_hooks.on_tool_start(context_wrapper, agent, tool)
        # ... other events
```
